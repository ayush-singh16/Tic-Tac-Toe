# <center style="font-family: consolas; font-size: 32px; font-weight: bold;">(‡≤†‡≤ø‚Å†_‚Å†‡≤†) Introduction</center>
### Tic-Tac-Toe(also known as noughts and crosses or Xs and Os) is a paper-and-pencil game for two players, X and O, who take turns marking the spaces in a 3√ó3 grid. The player who succeeds in placing three of their marks in a horizontal, vertical, or diagonal row wins the game.Given is the some experiences of winning or losing a game when the the player has opted for the moves at each positions in the grid. The total positions being 9. The chance of having values at each column can be X or O or b for blank

## <center style="font-family: consolas; font-size: 32px; font-weight: bold;">Goal:- Predict the Outcome of the game </center>

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings("ignore")
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
```

<a id="2"></a>
# <div style="box-shadow: rgb(60, 121, 245) 0px 0px 0px 3px inset, rgb(255, 255, 255) 10px -10px 0px -3px, rgb(31, 193, 27) 10px -10px, rgb(255, 255, 255) 20px -20px 0px -3px, rgb(255, 217, 19) 20px -20px, rgb(255, 255, 255) 30px -30px 0px -3px, rgb(255, 156, 85) 30px -30px, rgb(255, 255, 255) 40px -40px 0px -3px, rgb(255, 85, 85) 40px -40px; padding:20px; margin-right: 40px; font-size:30px; font-family: consolas; text-align:center; display:fill;"><b> 2. Data Collection & Processing</b></div>

```python
train_data = pd.read_csv('/kaggle/input/pesumcadatathon23/train.csv')
test_data = pd.read_csv('/kaggle/input/pesumcadatathon23/test.csv')
```

```python
train_data
```

```python
test_data
```

```python

def preprocess_data(data):
    data = data.replace({'x': 1, 'o': -1, 'b': 0})
    return data

train_data = preprocess_data(train_data)
test_data = preprocess_data(test_data)
```

# <center style="font-family: consolas; font-size: 28px; font-weight: bold;">Extract features (moves) from train data </center>

```python

X_train = train_data.drop(columns=['class']).values  
X_test = test_data.values
```

## <center style="font-family: consolas; font-size: 28px; font-weight: bold;">Apply K-Means clustering </center> 

```python

num_clusters = 4  # You can adjust the number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(X_train)
```

## <center style="font-family: consolas; font-size: 28px; font-weight: bold;">Separate clustered data</center>

```python
clustered_data = [X_train[clusters == cluster] for cluster in range(num_clusters)]
```

<a id="3"></a>
# <div style="box-shadow: rgb(60, 121, 245) 0px 0px 0px 3px inset, rgb(255, 255, 255) 10px -10px 0px -3px, rgb(31, 193, 27) 10px -10px, rgb(255, 255, 255) 20px -20px 0px -3px, rgb(255, 217, 19) 20px -20px, rgb(255, 255, 255) 30px -30px 0px -3px, rgb(255, 156, 85) 30px -30px, rgb(255, 255, 255) 40px -40px 0px -3px, rgb(255, 85, 85) 40px -40px; padding:20px; margin-right: 40px; font-size:30px; font-family: consolas; text-align:center; display:fill;"><b> 3. Calculate silhouette score to evaluate clustering quality</b></div>

```python
silhouette_avg = silhouette_score(X_train, clusters)
print("Silhouette Score:", silhouette_avg)

# Predict outcomes for test data based on clusters
y_pred = []
for x_test in X_test:
    cluster = kmeans.predict([x_test])[0]
    cluster_data = clustered_data[cluster]
    # Use the majority outcome from the cluster as the prediction
    outcome_counts = np.bincount(cluster_data[:, -1].astype(int))
    predicted_outcome = np.argmax(outcome_counts)
    y_pred.append(predicted_outcome)

# Convert predictions to numpy array
y_pred = np.array(y_pred)

# Print predicted outcomes
print("Predicted Outcomes:", y_pred)
```

<a id="4"></a>
# <div style="box-shadow: rgb(60, 121, 245) 0px 0px 0px 3px inset, rgb(255, 255, 255) 10px -10px 0px -3px, rgb(31, 193, 27) 10px -10px, rgb(255, 255, 255) 20px -20px 0px -3px, rgb(255, 217, 19) 20px -20px, rgb(255, 255, 255) 30px -30px 0px -3px, rgb(255, 156, 85) 30px -30px, rgb(255, 255, 255) 40px -40px 0px -3px, rgb(255, 85, 85) 40px -40px; padding:20px; margin-right: 40px; font-size:30px; font-family: consolas; text-align:center; display:fill;"><b> 4. Submission</b></div>

```python
my_submission = pd.DataFrame({'Id': test_data.Id,
                              'class': y_pred})


# Save submission DataFrame to a CSV file
my_submission.to_csv('/kaggle/working/submission.csv', index=False)
```
